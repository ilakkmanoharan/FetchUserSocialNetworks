```python
#K-Means Clustering:

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('Data/x.csv')

# Select the 'LABEL' attribute for clustering
X = data[['LABEL']]

# Define the number of clusters
k = 2

# Create a K-Means object and fit the data
kmeans = KMeans(n_clusters=k, random_state=0).fit(X)

# Perform K-Means clustering
#kmeans = KMeans(n_clusters=2)
#kmeans.fit(data.drop('index', axis=1))

# Get the labels for each node
labels = kmeans.labels_

# Add the labels to the original data
data['kmeans_labels'] = labels

# Create a scatter plot of the clusters
plt.scatter(data['index'], data['LABEL'], c=data['kmeans_labels'])
plt.title('K-Means Clustering')
plt.xlabel('Node Index')
plt.ylabel('LABEL')
plt.show()

# Create a heatmap of the clusters
sns.heatmap(pd.crosstab(data['LABEL'], data['kmeans_labels']), cmap="YlGnBu")
plt.title('K-Means Clustering')
plt.xlabel('Cluster')
plt.ylabel('LABEL')
plt.show()

# Evaluate the clustering performance using the silhouette score
silhouette_avg = silhouette_score(data.drop('index', axis=1), labels)
print("The average silhouette_score is :", silhouette_avg)

#The silhouette score ranges from -1 to 1, with a higher value indicating better clustering performance.

# Evaluate the clustering performance using the ARI and NMI
true_labels = data['LABEL']
ari = adjusted_rand_score(true_labels, labels)
nmi = normalized_mutual_info_score(true_labels, labels)
print("ARI: %.3f" % ari)
print("NMI: %.3f" % nmi)

#The ARI and NMI both range from 0 to 1, with a higher value indicating better clustering performance.
```

```python
#Hierarchical Clustering:

import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering

# Load the data
data = pd.read_csv('Data/x.csv')

# Extract the features
X = data.drop(['index', 'LABEL'], axis=1)

# Perform hierarchical clustering
hierarchical = AgglomerativeClustering(n_clusters=3)
labels = hierarchical.fit_predict(X)

# Add the labels to the original data
data['hierarchical_labels'] = labels

# Create a scatter plot of the clusters
plt.scatter(data['index'], data['LABEL'], c=data['hierarchical_labels'])
plt.title('Hierarchical Clustering')
plt.xlabel('Node Index')
plt.ylabel('LABEL')
plt.show()

# Create a heatmap of the clusters
sns.heatmap(pd.crosstab(data['LABEL'], data['hierarchical_labels']), cmap="YlGnBu")
plt.title('Hierarchical Clustering')
plt.xlabel('Cluster')
plt.ylabel('LABEL')
plt.show()

# Evaluate the clustering performance using the silhouette score
silhouette_avg = silhouette_score(data.drop('index', axis=1), labels)
print("The average silhouette_score is :", silhouette_avg)

#The silhouette score ranges from -1 to 1, with a higher value indicating better clustering performance.

# Evaluate the clustering performance using the ARI and NMI
true_labels = data['LABEL']
ari = adjusted_rand_score(true_labels, labels)
nmi = normalized_mutual_info_score(true_labels, labels)
print("ARI: %.3f" % ari)
print("NMI: %.3f" % nmi)

#The ARI and NMI both range from 0 to 1, with a higher value indicating better clustering performance.
```

```python
#DBSCAN:

import pandas as pd
from sklearn.cluster import DBSCAN

# Load the data
data = pd.read_csv('Data/x.csv')

# Select the 'LABEL' attribute for clustering
X = data[['LABEL']]

# Create a DBSCAN object and fit the data
dbscan = DBSCAN(eps=0.5, min_samples=5).fit(X)

# Get the labels for each node
labels = dbscan.labels_

# Add the labels to the original data
data['dbscan_labels'] = labels

# Create a scatter plot of the clusters
plt.scatter(data['index'], data['LABEL'], c=data['dbscan_labels'])
plt.title('DBSCAN Clustering')
plt.xlabel('Node Index')
plt.ylabel('LABEL')
plt.show()

# Create a heatmap of the clusters
sns.heatmap(pd.crosstab(data['LABEL'], data['dbscan_labels']), cmap="YlGnBu")
plt.title('DBSCAN Clustering')
plt.xlabel('Cluster')
plt.ylabel('LABEL')
plt.show()

# Evaluate the clustering performance using the silhouette score
silhouette_avg = silhouette_score(data.drop('index', axis=1), labels)
print("The average silhouette_score is :", silhouette_avg)

#The silhouette score ranges from -1 to 1, with a higher value indicating better clustering performance.

# Evaluate the clustering performance using the ARI and NMI
true_labels = data['LABEL']
ari = adjusted_rand_score(true_labels, labels)
nmi = normalized_mutual_info_score(true_labels, labels)
print("ARI: %.3f" % ari)
print("NMI: %.3f" % nmi)

#The ARI and NMI both range from 0 to 1, with a higher value indicating better clustering performance.
```

```python
#evaluate the clustering performance using the silhouette score:

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the data
data = pd.read_csv('Data/x.csv')

# Perform K-Means clustering
kmeans = KMeans(n_clusters=2)
kmeans.fit(data.drop('index', axis=1))

# Evaluate the clustering performance using the silhouette score
silhouette_avg = silhouette_score(data.drop('index', axis=1), kmeans.labels_)
print("The average silhouette_score is :", silhouette_avg)

#Here, we are using the K-Means algorithm to cluster the data into 2 clusters, 
#and then using the silhouette score to evaluate the quality of the resulting clusters. 
#The silhouette score ranges from -1 to 1, with a higher value indicating better clustering performance.
```

```python
#Evaluate the clustering performance using the Adjusted Rand Index (ARI) and Normalized Mutual Information (NMI):

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

# Load the data
data = pd.read_csv('Data/x.csv')

# Perform K-Means clustering
kmeans = KMeans(n_clusters=800)
kmeans.fit(data.drop('index', axis=1))

# Evaluate the clustering performance using the ARI and NMI
true_labels = data['LABEL']
ari = adjusted_rand_score(true_labels, kmeans.labels_)
nmi = normalized_mutual_info_score(true_labels, kmeans.labels_)
print("ARI: %.3f" % ari)
print("NMI: %.3f" % nmi)

#The ARI and NMI both range from 0 to 1, with a higher value indicating better clustering performance.
```

```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans

# Load the data
data = pd.read_csv('Data/x.csv')

# Perform K-Means clustering
kmeans = KMeans(n_clusters=2)
kmeans.fit(data.drop('index', axis=1))

# Create a heatmap of the clustering results
sns.heatmap(pd.concat([data['LABEL'], pd.Series(kmeans.labels_, name='cluster')], axis=1).groupby(['LABEL', 'cluster']).size().unstack(level=1).fillna(0), cmap='YlGnBu')
plt.xlabel('Cluster')
plt.ylabel('LABEL')
plt.title('K-Means Clustering Results')
plt.show()
```
